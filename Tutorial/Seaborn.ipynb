# Seaborn Tutorial 

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
fifa_data = pd.read_csv("dataforsns/fifa.csv", index_col="Date", parse_dates=True)
```

```python
fifa_data.head(5) # Also tail() for last rows of data
```

# Line Graph 

```python
sns.lineplot(data=fifa_data) # Ugly graph , congested
```

```python
plt.figure(figsize=(16,6)) # Yeah more space!
plt.title("Ranking of Countries at FIFA")
sns.lineplot(data=fifa_data)
```

```python
spotify_data = pd.read_csv("dataforsns/spotify.csv",index_col="Date",parse_dates=True)
```

###### Seperated Line Graphs

```python
flight_data = pd.read_csv("dataforsns/flight_delays.csv",index_col="Month")
```

```python
flight_data
```

# Bar Graph 

```python
plt.figure(figsize=(10,6))
sns.barplot(x=flight_data.index,y=flight_data['NK'])
plt.ylabel("Arrival delay (in min)")
```

# Heatmap

```python
plt.figure(figsize=(14,7))
sns.heatmap(data=flight_data,annot=True)
plt.xlabel("Airline")
```

```python
insurance_data = pd.read_csv("dataforsns/insurance.csv")
```

```python
insurance_data.head()
```

# Scatter Plot

```python
sns.scatterplot(x=insurance_data['bmi'],y=insurance_data['charges'])
```

###### Scatter Plot with Regression lines

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

###### Scatter Plot (3 variable)

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

###### Scatter Plot (3 variable) with Two Regression lines

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

###### Categorical Scatter Plot 

```python
sns.swarmplot(x=insurance_data['smoker'],y=insurance_data['charges'])
```
